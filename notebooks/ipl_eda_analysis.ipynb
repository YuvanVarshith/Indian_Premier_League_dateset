Exploratory Data Analysis (EDA) Questions



Match-Level Analysis


```python
#How many matches were played in each IPL season?

a= df.season.unique()
for lp in a:
    count= df[df['season']==lp].id.count()
    print(f'In {lp} {count} matches played')
```

```python
#Which teams have won the most number of matches
df["winner"].value_counts().head(1)
```

```python
# Which venue hosted the most matches?
df["venue"].value_counts().head(1)
```

```python
#How many times did teams win the toss and also win the match?
df[(df.toss_winner == df.winner)].id.count()
```

```python
#Who are the top 10 players with the most "Player of the Match" awards?
df["player_of_match"].value_counts().head(10)
```

```python
# Identify the most successful team for each individual season
a = df['season'].unique()
for lp in a:
    season= df[df['season']==lp]
    winner = season['winner'].value_counts().head(1)
    print(f"In {lp} season {winner.index[0]} has the most wins")
```

Ball-by-Ball Analysis


Who are the top 10 highest run scorers?

```python
batsmans = df1.batsman.unique()
for batsman_name in batsmans:
    batsman_total_runs  = df1.groupby('batsman')['total_runs'].sum()
    batsman_total_runs_sorted = batsman_total_runs.sort_values(ascending=False)
print(batsman_total_runs_sorted.head(10))
   
```
