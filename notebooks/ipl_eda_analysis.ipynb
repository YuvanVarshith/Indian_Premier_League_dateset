Exploratory Data Analysis (EDA) Questions



Match-Level Analysis


```python
#How many matches were played in each IPL season?

# Count matches per season
season_match_counts = df.groupby('season')['id'].count().sort_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(season_match_counts.index.astype(str), season_match_counts.values, color='skyblue')
plt.title('Number of Matches Played in Each IPL Season', fontsize=14)
plt.xlabel('IPL Season (Year)', fontsize=12)
plt.ylabel('Number of Matches', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
#Which teams have won the most number of matches
team_wins= df["winner"].value_counts().reset_index()
team_wins.columns = ['team', 'wins']


# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=team_wins, x='wins', y='team', palette='viridis')
plt.title("Most IPL Matches Won by Teams", fontsize=14)
plt.xlabel("Number of Wins")
plt.ylabel("Team")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Which venue hosted the most matches?
venue = df["venue"].value_counts().head(1)
venue_name = venue.index[0]
venue_count = venue.values[0]

plt.figure(figsize=(6, 4))
plt.text(0.5, 0.6, f"{venue_name}", fontsize=18, ha='center', weight='bold', color='darkblue')
plt.text(0.5, 0.4, f"{venue_count} Matches", fontsize=14, ha='center')
plt.axis('off')
plt.title("Most Frequent IPL Venue", fontsize=16)
plt.show()
```

```python
#How many times did teams win the toss and also win the match?
toss_match_wins =df[(df.toss_winner == df.winner)].id.count()

# Total matches played
total_matches = df.id.count()

# Values for pie chart
values = [toss_match_wins, total_matches - toss_match_wins]
labels = ['Toss & Match Won', 'Only Toss or Match Won']

# Plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140, colors=['mediumseagreen', 'lightcoral'])
plt.title('How Often Toss Winners Won the Match', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#Who are the top 10 players with the most "Player of the Match" awards?
top_players = df["player_of_match"].value_counts().head(10).reset_index()
top_players.columns = ['player', 'awards']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_players, y='player', x='awards', palette='plasma')
plt.title("Top 10 IPL Players with Most 'Player of the Match' Awards", fontsize=14)
plt.xlabel("Number of Awards")
plt.ylabel("Player")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Identify the most successful team for each individual season
mostwins_in_a_season  = []

for lp in df['season'].unique():
    season= df[df['season']==lp]
    winner = season['winner'].value_counts().head(1)
    mostwins_in_a_season.append({'season': lp, 'team': winner.index[0], 'wins': winner.values[0]})


season_winners_df = pd.DataFrame(mostwins_in_a_season).sort_values('season')

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=season_winners_df, x='season', y='wins', hue='team', dodge=False, palette='tab10')
plt.title("Most Successful Team in Each IPL Season", fontsize=14)
plt.xlabel("Season")
plt.ylabel("Number of Wins")
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Ball-by-Ball Analysis


```python
# Who are the top 10 highest run scorers?

batsmans = df1.groupby('batsman')['total_runs'].sum()
batsmans.sort_values(ascending=False).head(10)
# Get top 10 batsmen with highest runs
top_scorers = batsmans.sort_values(ascending=False).head(10).reset_index()
top_scorers.columns = ['batsman', 'total_runs']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_scorers, x='total_runs', y='batsman', hue='batsman', palette='inferno', legend=False)
plt.title("Top 10 Highest Run Scorers in IPL", fontsize=14)
plt.xlabel("Total Runs")
plt.ylabel("Batsman")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Which bowlers have taken the most wickets?

wickets_data = df1[(df1.dismissal_kind.notnull())&(~df1.dismissal_kind.isin(['run out', 'retired hurt','hit wicket','obstructing the field']))]
bowler_data = wickets_data.groupby('bowler').count()
bowler_data['player_dismissed'].sort_values(ascending=False).head(10)


# Get top 10 wicket-takers
top_bowlers = bowler_data['player_dismissed'].sort_values(ascending=False).head(10).reset_index()
top_bowlers.columns = ['bowler', 'wickets']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_bowlers, x='wickets', y='bowler', hue='bowler', palette='crest', legend=False)
plt.title("Top 10 Highest Wicket Takers in IPL", fontsize=14)
plt.xlabel("Wickets")
plt.ylabel("Bowler")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
#Which batsmen have the best strike rate (min 500 balls faced)?
batsmens_with_ball_faced= df1.groupby('batsman')['ball'].count()
batsmen_more_than_500_balls  =batsmens_with_ball_faced[batsmens_with_ball_faced>500]
list_of_batsmen_to_filter =   batsmen_more_than_500_balls.index.tolist()
runs_of_batsman_more_than_500_balls_faced= df1[df1['batsman'].isin(list_of_batsmen_to_filter)]
total_runs_scored = runs_of_batsman_more_than_500_balls_faced.groupby('batsman')['total_runs'].sum()
strike_rates = (total_runs_scored / batsmen_more_than_500_balls) * 100
best_strike_rates = strike_rates.sort_values(ascending=False).head(10)

best_strike_rates_df = best_strike_rates.reset_index()
best_strike_rates_df.columns = ['batsman', 'strike_rate']

plt.figure(figsize=(10, 6))
sns.barplot(data=best_strike_rates_df, x='strike_rate', y='batsman', hue='batsman', palette='magma', legend=False)
plt.title("Top 10 Batsmen with Best Strike Rates (Min 500 Balls Faced)", fontsize=14)
plt.xlabel("Strike Rate")
plt.ylabel("Batsman")
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add strike rate values on the bars
for index, row in best_strike_rates_df.iterrows():
    plt.text(row['strike_rate'] + 1, index, f"{row['strike_rate']:.2f}", va='center', fontsize=10)

plt.tight_layout()
plt.show()

```

```python
# Which bowler has the best economy rate (min 200 overs bowled)?
bowler_balls_count= df1.groupby('bowler')['ball'].count()
bowler_bowled_more_than_200_overs = bowler_balls_count[bowler_balls_count>1200]
list_of_bowlers_to_filter = bowler_bowled_more_than_200_overs.index.tolist()
bowler_data = df1[df1['bowler'].isin(list_of_bowlers_to_filter)]
runs_conceded_by_bowler = bowler_data.groupby('bowler')['total_runs'].sum()
economy_rate = (runs_conceded_by_bowler/(bowler_bowled_more_than_200_overs/6))
best_economy_rate = economy_rate.sort_values(ascending=True).head(10)
best_economy_rate

# Prepare DataFrame
best_economy_df = best_economy_rate.reset_index()
best_economy_df.columns = ['bowler', 'economy_rate']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=best_economy_df, x='economy_rate', y='bowler',hue='bowler', palette='viridis',legend=False)
plt.title("Top 10 Bowlers with Best Economy Rate (Min 200 Overs)", fontsize=14)
plt.xlabel("Economy Rate")
plt.ylabel("Bowler")
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add economy rate values to the bars
for index, row in best_economy_df.iterrows():
    plt.text(row['economy_rate'] + 0.05, index, f"{row['economy_rate']:.2f}", va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# What is the distribution of different dismissal types (caught, bowled, lbw, etc.)?

# Step 1: Get the dismissal type distribution
dismissals = df1['dismissal_kind'].value_counts()


# Step 2: Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(dismissals.values, labels=dismissals.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Dismissal Type Distribution in IPL")
plt.tight_layout()
plt.show()

```
